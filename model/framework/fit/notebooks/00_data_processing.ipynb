```python
import pandas as pd
import os

datapath = "../data"

activities = ['antiinfective','antiinflammatory','antineoplastic','cardio',
              'cns','dermatologic','gastrointestinal','hematologic',
              'lipidregulating','reproductivecontrol','respiratorysystem','urological']
```

```python
for activity in activities:

    # Get actives
    actives = sorted(set(pd.read_csv(os.path.join(datapath, f"{activity}_smiles_rmsaltol.csv"), header=None)[0].tolist()))

    # Get inactives
    inactives = []
    for a in activities:
        if a != activity:
            inactives.extend(sorted(set(pd.read_csv(os.path.join(datapath, f"{a}_smiles_rmsaltol.csv"), header=None)[0].tolist())))
    inactives_unfilt = set(inactives)
    inactives = sorted(set([i for i in inactives if i not in actives]))

    print(activity, len(actives), len(inactives_unfilt), len(inactives))

    df = pd.DataFrame({'smiles': actives + inactives,
                       'bin': [1]*len(actives) + [0]*len(inactives)})
    df.to_csv(os.path.join(datapath, "ml_datasets", f"{activity}.csv"), index=False)
```
